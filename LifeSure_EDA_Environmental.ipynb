This notebook explores four cleaned datasets to uncover insights on vehicle emissions, car and home insurance behaviors, and synthetic customer profiles.

Datasets used:
1. **Car CO2 Emissions** ‚Äì Environmental footprint of different vehicles
2. **Car Insurance** ‚Äì Risk and usage trends by vehicle type and claim history
3. **Home Insurance** ‚Äì Claims behavior by property type and location
4. **Synthetic Insurance** ‚Äì Demographic and behavioral traits of insurance customers

Each section includes:
- Key metrics
- Visual exploration
- Environmental and strategic insights


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

## Load Cleaned Datasets

```python
car_co2 = pd.read_csv("cleaned_car_co2.csv")
car_insurance = pd.read_csv("cleaned_car_insurance.csv")
home_insurance = pd.read_csv("cleaned_home_insurance.csv")
synthetic = pd.read_csv("cleaned_synthetic_insurance_data.csv")
```

## Dataset 1: Car CO2 Emissions


**Goal**: Identify which vehicle characteristics contribute most to carbon emissions and fuel consumption.  
**Business Use**: Targeting low-emission vehicles for eco-insurance discounts.


```python
sns.boxplot(data=car_co2, x='Fuel Type', y='CO2 Emissions(g/km)', palette='Reds')
plt.title("CO2 Emissions by Fuel Type")
plt.ylabel("CO2 Emissions (g/km)")
plt.xlabel("Fuel Type")
plt.show()
```

```python
sns.scatterplot(data=car_co2, x='Engine Size(L)', y='CO2 Emissions(g/km)', hue='Fuel Type', palette='Set1')
plt.title("CO2 Emissions vs Engine Size")
plt.xlabel("Engine Size (Liters)")
plt.ylabel("CO2 Emissions (g/km)")
plt.show()
```

```python
sns.barplot(data=car_co2, x='Vehicle Class', y='CO2 Emissions(g/km)', estimator='mean', palette='Blues')
plt.title("Average CO2 Emissions by Vehicle Class")
plt.xlabel("Vehicle Class")
plt.ylabel("Avg CO2 Emissions (g/km)")
plt.xticks(rotation=45)
plt.show()
```

## Dataset 2: Car Insurance


**Goal**: Assess relationships between vehicle features and insurance risk (claim rate, premium, insured value).  
**Business Use**: Combine risk and environmental data for green vehicle risk scoring.


```python
sns.boxplot(data=car_insurance, x='INSR_TYPE', y='PREMIUM', palette='Purples')
plt.title("Premiums by Insurance Type")
plt.xlabel("Insurance Type")
plt.ylabel("Premium (Currency)")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.histplot(data=car_insurance, x='CLAIM_PAID', bins=2)
plt.title("Distribution of Claims Paid")
plt.xlabel("Claim Paid (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
sns.barplot(data=car_insurance, x='TYPE_VEHICLE', y='CLAIM_PAID', estimator='mean', palette='coolwarm')
plt.title("Average Claim Rate by Vehicle Type")
plt.xlabel("Vehicle Type")
plt.ylabel("Average Claim Rate")
plt.xticks(rotation=30)
plt.show()
```

## Dataset 3: Home Insurance


**Goal**: Explore claim patterns in relation to building type and environmental features.  
**Business Use**: Promote sustainability incentives based on claim likelihood.


```python
sns.countplot(data=home_insurance, x='Building_Type', hue='Claim', palette='Set2')
plt.title("Claims by Building Type")
plt.xlabel("Building Type")
plt.ylabel("Number of Claims")
plt.xticks(rotation=30)
plt.show()
```

## Dataset 4: Synthetic Insurance Customers


**Goal**: Identify high-potential and environmentally-influenced customer segments.  
**Business Use**: Guide policy innovation and eco-targeted marketing.


```python
sns.histplot(synthetic['Claims_Frequency'], bins=10, kde=True, color='orange')
plt.title("Distribution of Claims Frequency")
plt.xlabel("Claims Frequency")
plt.ylabel("Customer Count")
plt.show()
```

```python
sns.scatterplot(data=synthetic, x='Premium_Amount', y='Total_Discounts', hue='Policy_Type', palette='Spectral')
plt.title("Total Discounts vs Premium Amount")
plt.xlabel("Premium Amount")
plt.ylabel("Total Discounts")
plt.show()
```

```python
sns.boxplot(data=synthetic, x='Region', y='Credit_Score', palette='coolwarm')
plt.title("Credit Score by Region")
plt.xlabel("Region")
plt.ylabel("Credit Score")
plt.xticks(rotation=45)
plt.show()
```
